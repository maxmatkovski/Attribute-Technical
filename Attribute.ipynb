```python
import pandas as pd
import numpy as np

# Load the data
accounts = pd.read_csv('accounts.csv')
billing = pd.read_csv('billing.csv')
usage = pd.read_csv('usage.csv')

print("="*80)
print("ACCOUNTS DATA STRUCTURE")
print("="*80)
print(f"\nShape: {accounts.shape}")
print(f"Columns: {accounts.columns.tolist()}")
print("\nFirst 5 rows:")
print(accounts.head())
print("\nData types:")
print(accounts.dtypes)
print("\nBasic statistics:")
print(accounts.describe(include='all'))
print("\nMissing values:")
print(accounts.isnull().sum())

print("\n\n")
print("="*80)
print("BILLING DATA STRUCTURE")
print("="*80)
print(f"\nShape: {billing.shape}")
print(f"Columns: {billing.columns.tolist()}")
print("\nFirst 5 rows:")
print(billing.head())
print("\nData types:")
print(billing.dtypes)
print("\nBasic statistics:")
print(billing.describe())
print("\nMissing values:")
print(billing.isnull().sum())

print("\n\n")
print("="*80)
print("USAGE DATA STRUCTURE")
print("="*80)
print(f"\nShape: {usage.shape}")
print(f"Columns: {usage.columns.tolist()}")
print("\nFirst 5 rows:")
print(usage.head())
print("\nData types:")
print(usage.dtypes)
print("\nBasic statistics:")
print(usage.describe())
print("\nMissing values:")
print(usage.isnull().sum())

print("\n\n")
print("="*80)
print("QUICK SUMMARY")
print("="*80)
print(f"Total accounts: {accounts.shape[0]}")
print(f"Total billing records: {billing.shape[0]}")
print(f"Total usage records: {usage.shape[0]}")
print(f"\nUnique accounts in billing: {billing['account_id'].nunique()}")
print(f"Unique accounts in usage: {usage['account_id'].nunique()}")
```

```python
# Step 3: Merge datasets and create key metrics

# Merge usage and billing
merged = usage.merge(billing, on=['account_id', 'month'], how='inner')
print(f"After merging usage + billing: {merged.shape}")

# Merge with accounts
df = merged.merge(accounts, on='account_id', how='left')
print(f"Final merged dataset: {df.shape}")

# Convert month to datetime
df['month'] = pd.to_datetime(df['month'])

# Create key business metrics
df['net_revenue'] = df['cost_after_discount_usd']  # Revenue after discounts
df['profit_margin'] = ((df['billed_after_discount_usd'] - df['cost_after_discount_usd']) / 
                       df['billed_after_discount_usd'] * 100)
df['total_usage'] = df['compute_hours'] + df['storage_gb'] + df['network_gb']

print("\n" + "="*80)
print("MERGED DATASET")
print("="*80)
print(f"Shape: {df.shape}")
print(f"Columns: {df.columns.tolist()}")
print("\nFirst few rows with key metrics:")
print(df[['account_id', 'month', 'plan_type', 'cost_usd', 'discount_usd', 
          'net_revenue', 'profit_margin']].head(10))

print("\n" + "="*80)
print("KEY BUSINESS METRICS")
print("="*80)
print(f"Total Revenue (after discounts): ${df['net_revenue'].sum():,.2f}")
print(f"Total Billed (after discounts): ${df['billed_after_discount_usd'].sum():,.2f}")
print(f"Total Discounts Given: ${df['discount_usd'].sum():,.2f}")
print(f"Average Discount Rate: {df['discount_percent'].mean():.2f}%")
print(f"Average Monthly Revenue per Customer: ${df.groupby('account_id')['net_revenue'].sum().mean():,.2f}")
print(f"Time Period: {df['month'].min()} to {df['month'].max()}")
print(f"Number of Months: {df['month'].nunique()}")
```

```python
# VISUALIZATION 1: Profitability & Growth - Revenue by Plan Type

import matplotlib.pyplot as plt
import seaborn as sns

# Aggregate by plan type
plan_metrics = df.groupby('plan_type').agg({
    'net_revenue': 'sum',
    'account_id': 'nunique',
    'discount_usd': 'sum',
    'profit_margin': 'mean'
}).reset_index()

plan_metrics.columns = ['plan_type', 'total_revenue', 'num_customers', 'total_discounts', 'avg_profit_margin']
plan_metrics['avg_revenue_per_customer'] = plan_metrics['total_revenue'] / plan_metrics['num_customers']

# Sort by revenue
plan_metrics = plan_metrics.sort_values('total_revenue', ascending=False)

print("Revenue by Plan Type:")
print(plan_metrics)
print()

# Create visualization
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Plot 1: Total Revenue and Customer Count by Plan Type
ax1 = axes[0]
x = range(len(plan_metrics))
width = 0.35

# Bar chart for revenue
bars1 = ax1.bar([i - width/2 for i in x], plan_metrics['total_revenue'], 
                width, label='Total Revenue', color='#2ecc71', alpha=0.8)

# Twin axis for customer count
ax1_twin = ax1.twinx()
bars2 = ax1_twin.bar([i + width/2 for i in x], plan_metrics['num_customers'], 
                     width, label='Customers', color='#3498db', alpha=0.8)

ax1.set_xlabel('Plan Type', fontsize=12, fontweight='bold')
ax1.set_ylabel('Total Revenue ($)', fontsize=11, color='#2ecc71', fontweight='bold')
ax1_twin.set_ylabel('Number of Customers', fontsize=11, color='#3498db', fontweight='bold')
ax1.set_title('Revenue & Customer Distribution by Plan Type', fontsize=14, fontweight='bold', pad=20)
ax1.set_xticks(x)
ax1.set_xticklabels(plan_metrics['plan_type'], fontsize=11)
ax1.tick_params(axis='y', labelcolor='#2ecc71')
ax1_twin.tick_params(axis='y', labelcolor='#3498db')
ax1.grid(axis='y', alpha=0.3)

# Add value labels on bars
for i, v in enumerate(plan_metrics['total_revenue']):
    ax1.text(i - width/2, v, f'${v/1000:.0f}K', ha='center', va='bottom', fontsize=9, fontweight='bold')
for i, v in enumerate(plan_metrics['num_customers']):
    ax1_twin.text(i + width/2, v, f'{v}', ha='center', va='bottom', fontsize=9, fontweight='bold')

# Legends
ax1.legend(loc='upper left', fontsize=10)
ax1_twin.legend(loc='upper right', fontsize=10)

# Plot 2: Average Revenue per Customer vs Profit Margin
ax2 = axes[1]
colors = ['#e74c3c', '#f39c12', '#9b59b6', '#1abc9c']
scatter = ax2.scatter(plan_metrics['avg_revenue_per_customer'], 
                     plan_metrics['avg_profit_margin'],
                     s=plan_metrics['num_customers']*2,
                     c=colors[:len(plan_metrics)],
                     alpha=0.6, edgecolors='black', linewidth=2)

# Add labels for each point
for idx, row in plan_metrics.iterrows():
    ax2.annotate(row['plan_type'], 
                (row['avg_revenue_per_customer'], row['avg_profit_margin']),
                xytext=(10, 10), textcoords='offset points',
                fontsize=11, fontweight='bold',
                bbox=dict(boxstyle='round,pad=0.5', facecolor='yellow', alpha=0.3))

ax2.set_xlabel('Avg Revenue per Customer ($)', fontsize=12, fontweight='bold')
ax2.set_ylabel('Avg Profit Margin (%)', fontsize=12, fontweight='bold')
ax2.set_title('Revenue vs Profitability by Plan Type\n(bubble size = customer count)', 
             fontsize=14, fontweight='bold', pad=20)
ax2.grid(True, alpha=0.3)
ax2.axhline(y=plan_metrics['avg_profit_margin'].mean(), color='red', 
           linestyle='--', linewidth=1, alpha=0.5, label='Avg Profit Margin')
ax2.legend(fontsize=10)

plt.tight_layout()
plt.savefig('viz1_profitability_by_plan.png', dpi=300, bbox_inches='tight')
plt.show()

print("\n✓ Visualization 1 saved as 'viz1_profitability_by_plan.png'")
```

```python
# VISUALIZATION 2: Growth Trends - Monthly Revenue & Customer Growth

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Monthly aggregation
monthly_trends = df.groupby('month').agg({
    'net_revenue': 'sum',
    'account_id': 'nunique',
    'cost_usd': 'sum',
    'discount_usd': 'sum'
}).reset_index()

monthly_trends.columns = ['month', 'revenue', 'active_customers', 'total_cost', 'total_discount']

# Calculate month-over-month growth
monthly_trends['revenue_growth_pct'] = monthly_trends['revenue'].pct_change() * 100
monthly_trends['customer_growth_pct'] = monthly_trends['active_customers'].pct_change() * 100

# Calculate cumulative metrics
monthly_trends['cumulative_revenue'] = monthly_trends['revenue'].cumsum()

print("Monthly Trends:")
print(monthly_trends[['month', 'revenue', 'active_customers', 'revenue_growth_pct', 'customer_growth_pct']].head(10))
print("\nLatest Month:")
print(monthly_trends[['month', 'revenue', 'active_customers', 'revenue_growth_pct']].tail(3))
print()

# Create visualization
fig, axes = plt.subplots(2, 1, figsize=(16, 10))

# Plot 1: Revenue and Active Customers Over Time
ax1 = axes[0]
ax1_twin = ax1.twinx()

# Revenue line
line1 = ax1.plot(monthly_trends['month'], monthly_trends['revenue'], 
                 marker='o', linewidth=2.5, markersize=8, color='#2ecc71', 
                 label='Monthly Revenue', markerfacecolor='white', markeredgewidth=2)

# Active customers line
line2 = ax1_twin.plot(monthly_trends['month'], monthly_trends['active_customers'], 
                      marker='s', linewidth=2.5, markersize=8, color='#3498db', 
                      label='Active Customers', markerfacecolor='white', markeredgewidth=2)

ax1.set_xlabel('Month', fontsize=12, fontweight='bold')
ax1.set_ylabel('Monthly Revenue ($)', fontsize=11, color='#2ecc71', fontweight='bold')
ax1_twin.set_ylabel('Active Customers', fontsize=11, color='#3498db', fontweight='bold')
ax1.set_title('Monthly Revenue & Customer Trends', fontsize=14, fontweight='bold', pad=20)
ax1.tick_params(axis='y', labelcolor='#2ecc71')
ax1_twin.tick_params(axis='y', labelcolor='#3498db')
ax1.grid(True, alpha=0.3)
ax1.tick_params(axis='x', rotation=45)

# Combine legends
lines = line1 + line2
labels = [l.get_label() for l in lines]
ax1.legend(lines, labels, loc='upper left', fontsize=11, framealpha=0.9)

# Highlight max revenue month
max_rev_idx = monthly_trends['revenue'].idxmax()
max_rev_month = monthly_trends.loc[max_rev_idx, 'month']
max_rev_value = monthly_trends.loc[max_rev_idx, 'revenue']
ax1.annotate(f'Peak: ${max_rev_value:,.0f}', 
            xy=(max_rev_month, max_rev_value),
            xytext=(20, 20), textcoords='offset points',
            bbox=dict(boxstyle='round,pad=0.5', facecolor='yellow', alpha=0.7),
            arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=0', lw=2),
            fontsize=10, fontweight='bold')

# Plot 2: Month-over-Month Growth Rates
ax2 = axes[1]
x = range(len(monthly_trends))
width = 0.35

# Filter out first month (no growth to calculate)
growth_data = monthly_trends[1:].copy()
growth_x = range(len(growth_data))

bars1 = ax2.bar([i - width/2 for i in growth_x], 
                growth_data['revenue_growth_pct'], 
                width, label='Revenue Growth %', 
                color=['#2ecc71' if x >= 0 else '#e74c3c' for x in growth_data['revenue_growth_pct']],
                alpha=0.7, edgecolor='black', linewidth=1)

bars2 = ax2.bar([i + width/2 for i in growth_x], 
                growth_data['customer_growth_pct'], 
                width, label='Customer Growth %', 
                color=['#3498db' if x >= 0 else '#e67e22' for x in growth_data['customer_growth_pct']],
                alpha=0.7, edgecolor='black', linewidth=1)

ax2.set_xlabel('Month', fontsize=12, fontweight='bold')
ax2.set_ylabel('Growth Rate (%)', fontsize=12, fontweight='bold')
ax2.set_title('Month-over-Month Growth Rates', fontsize=14, fontweight='bold', pad=20)
ax2.set_xticks(growth_x)
ax2.set_xticklabels([d.strftime('%Y-%m') for d in growth_data['month']], 
                     rotation=45, ha='right', fontsize=9)
ax2.axhline(y=0, color='black', linestyle='-', linewidth=1.5)
ax2.legend(fontsize=11, loc='upper left')
ax2.grid(True, alpha=0.3, axis='y')

# Add average growth line
avg_rev_growth = growth_data['revenue_growth_pct'].mean()
ax2.axhline(y=avg_rev_growth, color='#2ecc71', linestyle='--', 
           linewidth=2, alpha=0.5, label=f'Avg Rev Growth: {avg_rev_growth:.1f}%')

plt.tight_layout()
plt.savefig('viz2_growth_trends.png', dpi=300, bbox_inches='tight')
plt.show()

print("\n✓ Visualization 2 saved as 'viz2_growth_trends.png'")

# Summary statistics
print("\n" + "="*80)
print("GROWTH SUMMARY")
print("="*80)
print(f"Average Monthly Revenue Growth: {growth_data['revenue_growth_pct'].mean():.2f}%")
print(f"Average Customer Growth: {growth_data['customer_growth_pct'].mean():.2f}%")
print(f"Highest Revenue Growth: {growth_data['revenue_growth_pct'].max():.2f}% in {growth_data.loc[growth_data['revenue_growth_pct'].idxmax(), 'month'].strftime('%Y-%m')}")
print(f"Months with Negative Revenue Growth: {(growth_data['revenue_growth_pct'] < 0).sum()}")
print(f"Total Revenue Growth (First to Last Month): {((monthly_trends['revenue'].iloc[-1] / monthly_trends['revenue'].iloc[0]) - 1) * 100:.2f}%")
```

```python
# VISUALIZATION 3: Usage Behavior - Resource Utilization Patterns

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Aggregate usage by plan type
usage_by_plan = df.groupby('plan_type').agg({
    'compute_hours': 'mean',
    'storage_gb': 'mean',
    'network_gb': 'mean',
    'active_users': 'mean',
    'net_revenue': 'mean'
}).reset_index()

print("Average Usage by Plan Type:")
print(usage_by_plan)
print()

# Calculate correlation between usage and revenue
usage_corr = df[['compute_hours', 'storage_gb', 'network_gb', 'active_users', 
                  'cost_usd', 'net_revenue', 'discount_percent']].corr()

print("Correlation Matrix:")
print(usage_corr)
print()

# Create visualization
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Plot 1: Average Resource Usage by Plan Type
ax1 = axes[0]
x = np.arange(len(usage_by_plan))
width = 0.25

bars1 = ax1.bar(x - width, usage_by_plan['compute_hours'], width, 
                label='Compute Hours', color='#e74c3c', alpha=0.8, edgecolor='black')
bars2 = ax1.bar(x, usage_by_plan['storage_gb'], width, 
                label='Storage GB', color='#3498db', alpha=0.8, edgecolor='black')
bars3 = ax1.bar(x + width, usage_by_plan['network_gb'], width, 
                label='Network GB', color='#f39c12', alpha=0.8, edgecolor='black')

ax1.set_xlabel('Plan Type', fontsize=12, fontweight='bold')
ax1.set_ylabel('Average Usage', fontsize=12, fontweight='bold')
ax1.set_title('Average Resource Usage by Plan Type', fontsize=14, fontweight='bold', pad=20)
ax1.set_xticks(x)
ax1.set_xticklabels(usage_by_plan['plan_type'], fontsize=11)
ax1.legend(fontsize=11, loc='upper left')
ax1.grid(axis='y', alpha=0.3)

# Add value labels on bars
for i, (comp, stor, net) in enumerate(zip(usage_by_plan['compute_hours'], 
                                            usage_by_plan['storage_gb'], 
                                            usage_by_plan['network_gb'])):
    ax1.text(i - width, comp, f'{comp:.0f}', ha='center', va='bottom', fontsize=8, fontweight='bold')
    ax1.text(i, stor, f'{stor:.0f}', ha='center', va='bottom', fontsize=8, fontweight='bold')
    ax1.text(i + width, net, f'{net:.0f}', ha='center', va='bottom', fontsize=8, fontweight='bold')

# Plot 2: Correlation Heatmap
ax2 = axes[1]
# Select key metrics for heatmap
metrics_to_show = ['compute_hours', 'storage_gb', 'network_gb', 'active_users', 'net_revenue', 'discount_percent']
corr_subset = usage_corr.loc[metrics_to_show, metrics_to_show]

# Create heatmap
sns.heatmap(corr_subset, annot=True, fmt='.2f', cmap='coolwarm', center=0,
            square=True, ax=ax2, cbar_kws={'label': 'Correlation'},
            linewidths=1, linecolor='white',
            vmin=-1, vmax=1, cbar=True)

ax2.set_title('Usage & Revenue Correlation Matrix', fontsize=14, fontweight='bold', pad=20)
ax2.set_xticklabels(ax2.get_xticklabels(), rotation=45, ha='right', fontsize=9)
ax2.set_yticklabels(ax2.get_yticklabels(), rotation=0, fontsize=9)

plt.tight_layout()
plt.savefig('viz3_usage_behavior.png', dpi=300, bbox_inches='tight')
plt.show()

print("\n✓ Visualization 3 saved as 'viz3_usage_behavior.png'")

# Additional insights
print("\n" + "="*80)
print("USAGE BEHAVIOR INSIGHTS")
print("="*80)
print(f"\nStrongest correlation with revenue: {usage_corr['net_revenue'].abs().sort_values(ascending=False).index[1]} "
      f"({usage_corr['net_revenue'].abs().sort_values(ascending=False).values[1]:.3f})")
print(f"\nAverage resource usage across all customers:")
print(f"  - Compute Hours: {df['compute_hours'].mean():.2f} hours/month")
print(f"  - Storage: {df['storage_gb'].mean():.2f} GB/month")
print(f"  - Network: {df['network_gb'].mean():.2f} GB/month")
print(f"  - Active Users: {df['active_users'].mean():.2f} users/month")

# Calculate usage intensity by plan
usage_by_plan['total_usage'] = (usage_by_plan['compute_hours'] + 
                                 usage_by_plan['storage_gb'] + 
                                 usage_by_plan['network_gb'])
usage_by_plan['revenue_per_unit'] = usage_by_plan['net_revenue'] / usage_by_plan['total_usage']

print(f"\nRevenue efficiency ($/unit usage):")
for idx, row in usage_by_plan.iterrows():
    print(f"  - {row['plan_type']}: ${row['revenue_per_unit']:.2f} per unit usage")
```

```python
# VISUALIZATION 4: Risk & Opportunity - Revenue Concentration & Discount Analysis

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Calculate customer-level metrics
customer_metrics = df.groupby('account_id').agg({
    'net_revenue': 'sum',
    'discount_usd': 'sum',
    'cost_usd': 'sum',
    'month': 'count',
    'plan_type': 'first',
    'industry': 'first',
    'region': 'first'
}).reset_index()

customer_metrics.columns = ['account_id', 'total_revenue', 'total_discount', 
                            'total_cost', 'months_active', 'plan_type', 'industry', 'region']
customer_metrics['avg_discount_rate'] = (customer_metrics['total_discount'] / 
                                          customer_metrics['total_cost'] * 100).round(2)

# Sort by revenue for concentration analysis
customer_metrics_sorted = customer_metrics.sort_values('total_revenue', ascending=False).reset_index(drop=True)
customer_metrics_sorted['cumulative_revenue'] = customer_metrics_sorted['total_revenue'].cumsum()
customer_metrics_sorted['cumulative_revenue_pct'] = (customer_metrics_sorted['cumulative_revenue'] / 
                                                      customer_metrics_sorted['total_revenue'].sum() * 100)
customer_metrics_sorted['customer_rank_pct'] = ((customer_metrics_sorted.index + 1) / 
                                                 len(customer_metrics_sorted) * 100)

print("Top 10 Customers by Revenue:")
print(customer_metrics_sorted[['account_id', 'plan_type', 'industry', 'total_revenue', 
                                'avg_discount_rate']].head(10))
print()

# Discount analysis by plan
discount_by_plan = customer_metrics.groupby('plan_type').agg({
    'avg_discount_rate': ['mean', 'median', 'std']
}).reset_index()
discount_by_plan.columns = ['plan_type', 'mean_discount', 'median_discount', 'std_discount']

print("Discount Analysis by Plan Type:")
print(discount_by_plan)
print()

# Create visualization
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Plot 1: Revenue Concentration Curve (Lorenz Curve)
ax1 = axes[0]
ax1.plot(customer_metrics_sorted['customer_rank_pct'], 
         customer_metrics_sorted['cumulative_revenue_pct'], 
         linewidth=3, color='#e74c3c', label='Actual Revenue Distribution')
ax1.plot([0, 100], [0, 100], linestyle='--', color='gray', 
         linewidth=2, alpha=0.7, label='Perfect Equality')
ax1.fill_between(customer_metrics_sorted['customer_rank_pct'], 
                 0, customer_metrics_sorted['cumulative_revenue_pct'], 
                 alpha=0.2, color='#e74c3c')

ax1.set_xlabel('Cumulative % of Customers', fontsize=12, fontweight='bold')
ax1.set_ylabel('Cumulative % of Revenue', fontsize=12, fontweight='bold')
ax1.set_title('Revenue Concentration Analysis\n(Lorenz Curve)', 
             fontsize=14, fontweight='bold', pad=20)
ax1.grid(True, alpha=0.3)
ax1.legend(fontsize=11, loc='upper left')
ax1.set_xlim(0, 100)
ax1.set_ylim(0, 100)

# Add annotations for key percentiles
key_percentiles = [10, 20, 50, 80]
for pct in key_percentiles:
    idx = (customer_metrics_sorted['customer_rank_pct'] - pct).abs().idxmin()
    rev_pct = customer_metrics_sorted.loc[idx, 'cumulative_revenue_pct']
    ax1.plot([pct, pct], [0, rev_pct], 'k--', alpha=0.3, linewidth=1)
    ax1.plot([0, pct], [rev_pct, rev_pct], 'k--', alpha=0.3, linewidth=1)
    ax1.scatter([pct], [rev_pct], s=100, c='red', zorder=5, edgecolors='black', linewidth=2)
    ax1.annotate(f'Top {pct}%\n= {rev_pct:.0f}% rev', 
                xy=(pct, rev_pct), xytext=(pct-8, rev_pct+8),
                fontsize=9, fontweight='bold',
                bbox=dict(boxstyle='round,pad=0.4', facecolor='yellow', alpha=0.7))

# Plot 2: Discount Rate Distribution by Plan Type
ax2 = axes[1]

# Create box plot
plan_types = customer_metrics['plan_type'].unique()
discount_data = [customer_metrics[customer_metrics['plan_type'] == pt]['avg_discount_rate'].values 
                 for pt in plan_types]

bp = ax2.boxplot(discount_data, labels=plan_types, patch_artist=True,
                 widths=0.6, showmeans=True,
                 meanprops=dict(marker='D', markerfacecolor='red', markersize=8),
                 medianprops=dict(color='black', linewidth=2),
                 boxprops=dict(facecolor='#3498db', alpha=0.6, edgecolor='black', linewidth=1.5),
                 whiskerprops=dict(color='black', linewidth=1.5),
                 capprops=dict(color='black', linewidth=1.5),
                 flierprops=dict(marker='o', markerfacecolor='red', markersize=6, alpha=0.5))

ax2.set_xlabel('Plan Type', fontsize=12, fontweight='bold')
ax2.set_ylabel('Discount Rate (%)', fontsize=12, fontweight='bold')
ax2.set_title('Discount Rate Distribution by Plan Type\n(Diamond = Mean, Line = Median)', 
             fontsize=14, fontweight='bold', pad=20)
ax2.grid(axis='y', alpha=0.3)

# Add mean values as text
for i, pt in enumerate(plan_types):
    mean_val = customer_metrics[customer_metrics['plan_type'] == pt]['avg_discount_rate'].mean()
    ax2.text(i+1, mean_val, f'{mean_val:.1f}%', ha='center', va='bottom', 
            fontsize=10, fontweight='bold', color='red')

plt.tight_layout()
plt.savefig('viz4_risk_opportunity.png', dpi=300, bbox_inches='tight')
plt.show()

print("\n✓ Visualization 4 saved as 'viz4_risk_opportunity.png'")

# Risk & Opportunity Summary
print("\n" + "="*80)
print("RISK & OPPORTUNITY SUMMARY")
print("="*80)

# Revenue concentration risk
top_10_pct_customers = int(len(customer_metrics) * 0.1)
top_10_revenue = customer_metrics_sorted.head(top_10_pct_customers)['total_revenue'].sum()
top_20_pct_customers = int(len(customer_metrics) * 0.2)
top_20_revenue = customer_metrics_sorted.head(top_20_pct_customers)['total_revenue'].sum()
total_revenue = customer_metrics['total_revenue'].sum()

print(f"\nREVENUE CONCENTRATION RISK:")
print(f"  - Top 10% of customers generate: {top_10_revenue/total_revenue*100:.1f}% of revenue")
print(f"  - Top 20% of customers generate: {top_20_revenue/total_revenue*100:.1f}% of revenue")
print(f"  - Largest customer revenue: ${customer_metrics_sorted.iloc[0]['total_revenue']:,.2f} "
      f"({customer_metrics_sorted.iloc[0]['total_revenue']/total_revenue*100:.2f}% of total)")

print(f"\nDISCOUNT ANALYSIS:")
print(f"  - Average discount rate: {customer_metrics['avg_discount_rate'].mean():.2f}%")
print(f"  - Customers with >10% discount: {(customer_metrics['avg_discount_rate'] > 10).sum()} "
      f"({(customer_metrics['avg_discount_rate'] > 10).sum()/len(customer_metrics)*100:.1f}%)")
print(f"  - Total revenue lost to discounts: ${customer_metrics['total_discount'].sum():,.2f}")

# Industry concentration
industry_revenue = df.groupby('industry')['net_revenue'].sum().sort_values(ascending=False)
print(f"\nINDUSTRY CONCENTRATION:")
print(f"  - Top industry: {industry_revenue.index[0]} "
      f"({industry_revenue.iloc[0]/industry_revenue.sum()*100:.1f}% of revenue)")
print(f"  - Top 3 industries: {industry_revenue.head(3).sum()/industry_revenue.sum()*100:.1f}% of revenue")

# Churn risk indicators
low_usage_customers = df.groupby('account_id')['total_usage'].mean().sort_values().head(50)
print(f"\nPOTENTIAL CHURN RISK:")
print(f"  - Customers with lowest average usage: {len(low_usage_customers)} accounts identified")
print(f"  - These represent ${customer_metrics[customer_metrics['account_id'].isin(low_usage_customers.index)]['total_revenue'].sum():,.2f} in revenue")

print(f"\nOPPORTUNITIES:")
print(f"  - Free plan customers: {(customer_metrics['plan_type'] == 'Free').sum()} accounts")
print(f"  - Potential upgrade revenue: ${customer_metrics[customer_metrics['plan_type'] == 'Free']['total_revenue'].sum():,.2f} current revenue from Free tier")
print(f"  - Standard plan customers eligible for Pro/Enterprise: {(customer_metrics['plan_type'] == 'Standard').sum()} accounts")
```
